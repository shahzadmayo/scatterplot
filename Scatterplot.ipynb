# Creating a Scatter Plot

```python
# import the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Let’s visualize of “fmri” dataset using seaborn.scatterplot() function.

 
 
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")
sns.scatterplot(x="timepoint",
                    y="signal",
                    data=fmri)
fmri.head()
```

# Grouping data points on the basis of category, here as region and event.

```python
sns.scatterplot(x="timepoint",
                    y="signal",
                    hue="region",
                    style="event",
                     data=fmri)
```

# Basic visualization of “tips” dataset using Scatterplot

```python
tip = sns.load_dataset('tips')
sns.scatterplot(x='day', y='tip', data=tip)
tip.head()
```

# Grouping variables in Seaborn Scatter Plot with different attributes

```python

sns.scatterplot(x='day', y='tip', data= tip, marker = '+')
```

```python
# Adding the Hue attributes
sns.scatterplot(x='day', y='tip', data= tip,  hue='time')
```

```python
# Adding the Style attribute    
sns.scatterplot(x='day', y='tip', data=tip, hue="time", style="sex")
tip.head()
```

```python
# adding the platte attributes

sns.scatterplot(x='day', y='tip', data=tip, hue='time', palette='pastel')
```

```python
# adding the size attribute

sns.scatterplot(x='day', y='tip', data=tip ,hue='size', size = "size")

tip.head()
```

```python
#Adding legend attributes.
sns.scatterplot(x='day', y='tip', data=tip, hue='day',
					sizes=(30, 200), legend='full')
```

```python
# Adding alpha attributes.
sns.scatterplot(x='day', y='tip', data=tip, alpha = 0.1)

```

# How To Make Scatter Plot with Regression Line using Seaborn in Python?

## Plot the graph with the help of regplot() or lmplot() method.

```python

# load data
df = sns.load_dataset('iris')

# use regplot
sns.regplot(x = "sepal_length",
			y = "petal_length",
			ci = None,
			data = df)
```

```python


# use lmplot
sns.lmplot(x = "sepal_length",
			y = "petal_length",
			ci = None,
			data = df)
```
